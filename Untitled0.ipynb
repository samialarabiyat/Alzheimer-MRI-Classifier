```python
# 👇 Make sure to upload 'alzheimer_train.csv' and 'alzheimer_test.csv' before running

import pandas as pd
import numpy as np
from PIL import Image
import io
import ast
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from tqdm import tqdm

# Settings
IMG_SIZE = 128  # Target image size (128x128)
NUM_CLASSES = 4  # Number of classes (0 to 3)

# Load CSV files
train_df = pd.read_csv("alzheimer_train.csv")
test_df = pd.read_csv("alzheimer_test.csv")


```

```python
# Function to decode bytes into a grayscale image
def decode_image(row):
    img_dict = ast.literal_eval(row)
    byte_data = img_dict['bytes']
    image = Image.open(io.BytesIO(byte_data)).convert('L')  # Convert to grayscale
    image = image.resize((IMG_SIZE, IMG_SIZE))
    return np.array(image)
```

```python

# Convert image bytes to numpy arrays
X_train = np.stack([decode_image(row) for row in tqdm(train_df['image'])])
y_train = train_df['label'].values

X_test = np.stack([decode_image(row) for row in tqdm(test_df['image'])])
y_test = test_df['label'].values
```

```python
# Reshape and normalize images
X_train = X_train[..., np.newaxis] / 255.0
X_test = X_test[..., np.newaxis] / 255.0

# One-hot encode labels
lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_test = lb.transform(y_test)
```

```python
# Split training data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Build a simple CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 1)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dense(NUM_CLASSES, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()


```

```python

# Train the model
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=20, batch_size=32)

# Evaluate the model
loss, acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {acc:.4f}")
```

```python
# Predict on test data
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
# Change y_test_ohe to y_test
y_true_classes = np.argmax(y_test, axis=1)

# Confusion matrix
# You also need to import `confusion_matrix`, `matplotlib.pyplot` as `plt` and `seaborn` as `sns`
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

# Misclassified examples
wrong_indices = np.where(y_pred_classes != y_true_classes)[0]
print(f"Found {len(wrong_indices)} misclassified images")

# Show first 5 misclassified images
for i in wrong_indices[:5]:
    # Need to remove the extra dimension for displaying grayscale images with matplotlib
    plt.imshow(X_test[i].squeeze(), cmap='gray')
    plt.title(f"True: {y_true_classes[i]}, Predicted: {y_pred_classes[i]}")
    plt.axis('off')
    plt.show()
```

```python
import numpy as np
import pandas as pd
from PIL import Image
import io, ast
from tqdm import tqdm
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import confusion_matrix

# -------------------- Load & Preprocess Data --------------------
IMG_SIZE = 128
NUM_CLASSES = 4

# Load CSVs
train_df = pd.read_csv("alzheimer_train.csv")
test_df = pd.read_csv("alzheimer_test.csv")

# Decode image bytes
def decode_image(row):
    img_dict = ast.literal_eval(row)
    byte_data = img_dict['bytes']
    image = Image.open(io.BytesIO(byte_data)).convert('L')
    image = image.resize((IMG_SIZE, IMG_SIZE))
    return np.array(image)

# Convert to arrays
X_train = np.stack([decode_image(row) for row in tqdm(train_df['image'])])
y_train = train_df['label'].values
X_test = np.stack([decode_image(row) for row in tqdm(test_df['image'])])
y_test = test_df['label'].values

# Normalize and reshape
X_train = X_train[..., np.newaxis] / 255.0
X_test = X_test[..., np.newaxis] / 255.0

# One-hot encode labels
lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_test = lb.transform(y_test)

# Split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)

# -------------------- Build Model (Functional API) --------------------
inputs = Input(shape=(IMG_SIZE, IMG_SIZE, 1))
x = Conv2D(32, (3, 3), activation='relu')(inputs)
x = MaxPooling2D(2, 2)(x)
x = Conv2D(64, (3, 3), activation='relu', name="target_conv")(x)
x = MaxPooling2D(2, 2)(x)
x = Flatten()(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
outputs = Dense(NUM_CLASSES, activation='softmax')(x)

model = Model(inputs=inputs, outputs=outputs)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# -------------------- Train --------------------
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10)

# -------------------- Grad-CAM Utilities --------------------
def get_gradcam(model, img_array, class_idx, layer_name="target_conv"):
    grad_model = Model([model.inputs], [model.get_layer(layer_name).output, model.output])
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:, class_idx]

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(conv_outputs * pooled_grads, axis=-1)

    heatmap = np.maximum(heatmap, 0)
    heatmap /= tf.reduce_max(heatmap) + 1e-8
    return heatmap.numpy()

# -------------------- Predictions --------------------
y_pred = np.argmax(model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)
wrong_indices = np.where(y_pred != y_true)[0]
print(f"❌ Found {len(wrong_indices)} misclassified images")

# -------------------- Visualize Grad-CAM --------------------
for i in wrong_indices[:5]:
    img = X_test[i]
    true_label = y_true[i]
    pred_label = y_pred[i]

    img_input = np.expand_dims(img, axis=0).astype(np.float32)
    heatmap = get_gradcam(model, img_input, pred_label)

    heatmap_resized = cv2.resize(heatmap, (IMG_SIZE, IMG_SIZE))
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)

    img_rgb = np.repeat((img * 255).astype(np.uint8), 3, axis=2)
    superimposed_img = cv2.addWeighted(img_rgb, 0.6, heatmap_color, 0.4, 0)

    plt.imshow(superimposed_img)
    plt.title(f"True: {true_label}, Predicted: {pred_label}")
    plt.axis('off')
    plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Recompute predictions if needed
y_pred = np.argmax(model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)

# 🔍 Classification Report
print("📊 Classification Report:")
print(classification_report(y_true, y_pred, digits=4))

# 🔍 Confusion Matrix (raw counts)
cm = confusion_matrix(y_true, y_pred)

# 🔍 Confusion Matrix (percentages)
cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

# 📊 Visualization with seaborn (percent)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_percent, annot=True, fmt=".1f", cmap="Blues", cbar=True,
            xticklabels=[f"Pred {i}" for i in range(NUM_CLASSES)],
            yticklabels=[f"True {i}" for i in range(NUM_CLASSES)])
plt.title("Confusion Matrix (%)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```
